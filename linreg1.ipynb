```python
import numpy as np
import pandas as pd
from sklearn import linear_model
import matplotlib.pyplot as plt

linReg = linear_model.LinearRegression()
data = pd.read_csv('insurance.csv')
test = pd.read_csv('insuranceTest.csv')

X = data[['bmi']]  # Making X 2-dimensional by using double square brackets
y = data['charges']

x = test[['bmi']]  # Similar adjustment for the test data

linReg.fit(X, y)
p = linReg.predict(x)

plt.title('Charges v/s BMI')
plt.xlabel('BMI')
plt.ylabel('Chages')


plt.scatter(X,y)
plt.plot(x, p, color='red', linewidth=3)
plt.show()
```
